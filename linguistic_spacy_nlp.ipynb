##(1) Task => มองเป็น 1 Predicate
- Ex: Call the help desk

- Ex: make a request

##(1) [and] (2) => Composite Activity
- Ex: Call the help desk [and] make a request
- Ex: Pattern [VP][and][VP] => [Activity][and][Activity]

#ARTIFACT

##(1) Simple ARTIFACT หรือ <Activity> [Artifact]
- Ex: <Call> [the help desk] 

##(2) Complex Artifact หรือ [Activity][Artifact][and, or] [Artifact]
- Ex: [Contact customer via][email][or][telephone]
- Pattern : [NNP NN IN][NN][CC][NN]
